```python
#Linear Regression
import pandas as pd 
import numpy as np
data = pd.read_csv('train all Linear Regression.csv')
```

```python
# x = all columns - end column
# y = end column
x = data.iloc[:,:-1].values
y = data.iloc[:, 1].values
# كدا قسمت الداتا ل x , y ( feature , label)
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
data.head() 
# بيظهر اول 5 ريكورد
```

```python
data.shape
#الداتا فيها كام  صف و كام عموووود 
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train,y_train)
print("intercept",reg.intercept_)
print("cof",reg.coef_)
y_pred = reg.predict(x_test) # الي بسجل فيه الاجابة
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
print(df)
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
import matplotlib.pyplot as plt

data.plot(x='X', y='Y', style='^') # اسماء الاعمدة في الداتا بتاعتي 
plt.title('ْX vs Y')
plt.xlabel('X Studied')
plt.ylabel('Y Score')
plt.plot(x_test,y_pred, color="red", linewidth=3)
plt.show()
```
